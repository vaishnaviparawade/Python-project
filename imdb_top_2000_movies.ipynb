```python
import matplotlib.pyplot as plt
```

```python
# Create DataFrame
df = pd.DataFrame(data)
```

```python
plt.figure(figsize=(10, 6))
plt.hist(df['IMDB Rating'], bins=10, color='skyblue', edgecolor='black')
plt.xlabel('IMDB Rating')
plt.ylabel('Frequency')
plt.title('Distribution of IMDB Ratings')
plt.grid(True)
plt.show()
```

```python
# Plotting scatter plot for IMDB ratings and movie durations
plt.figure(figsize=(10, 6))
plt.scatter(df['Duration'], df['IMDB Rating'], color='skyblue', alpha=0.7)
plt.xlabel('Duration')
plt.ylabel('IMDB Rating')
plt.title('Scatter Plot of IMDB Ratings vs Movie Durations')
plt.grid(True)
plt.show()
```

```python
# Calculate the count of movies for each genre
genre_counts = df['Genre'].str.split(', ').explode().value_counts()

# Plotting pie chart for movie genres
plt.figure(figsize=(8, 8))
plt.pie(genre_counts, labels=genre_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Movie Genres')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```
